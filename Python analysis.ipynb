```python
##q1 Average Salary (USD) by Experience Level
avg_salary_exp = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()

# Now sort in ascending order of salary
avg_salary_exp = avg_salary_exp.sort_values(by='salary_in_usd', ascending=False)

print(avg_salary_exp.to_string(index=False))
```

```python
## q2 Find Top 5 Highest Paying Jobs and their location
top5_jobs = df[['job_title', 'salary_in_usd','company_location']].sort_values(by='salary_in_usd', ascending=False).head(5)
print(top5_jobs.to_string(index=False))
```

```python
## q3 Count the Number of Remote Jobs (remote_ratio = 100)
remote_jobs_count = df[df['remote_ratio'] == 100].shape[0]
print(remote_jobs_count)
```

```python
## q4 Average Salary by Company Size
avg_salary_company_size = df.groupby('company_size')['salary_in_usd'].mean().reset_index()
avg_salary_company_size = avg_salary_company_size.sort_values(by='salary_in_usd', ascending=False)
print(avg_salary_company_size.to_string(index=False))
```

```python
## q5 Find Highest Salary for Each Experience Level
highest_salary_per_experience = df.groupby('experience_level')['salary_in_usd'].max().reset_index()
highest_salary_per_experience = highest_salary_per_experience.sort_values(by='salary_in_usd', ascending=False)
print(highest_salary_per_experience.to_string(index=False))
```

```python
## q6 Average Salary by Company Size and Employment Type
avg_salary = df.groupby(['company_size', 'employment_type'])['salary_in_usd'].mean().reset_index()
avg_salary = avg_salary.sort_values(by='salary_in_usd', ascending=False)
print(avg_salary.to_string(index=False))
```

```python
## q7 Count Employees per Experience Level for Each Company Size
employee_count = (
    df.groupby(['company_size', 'experience_level'])
      .size()
      .reset_index(name='employee_count')
      .sort_values(by='employee_count', ascending=False)
)

print(employee_count.to_string(index=False))
```

```python
## q8 Top 5 Locations with Most Employees
top_locations = df['company_location'].value_counts().head(5).reset_index()
top_locations.columns = ['company_location', 'employee_count']
print(top_locations.to_string(index=False))
```

```python
## q9 Difference between Maximum and Minimum Salary per Experience Level
salary_gap = (df.groupby('experience_level')['salary_in_usd']
                .agg(lambda x: x.max() - x.min())
                .reset_index()
                .rename(columns={'salary_in_usd': 'salary_gap'}))
salary_gap = salary_gap.sort_values(by='salary_gap', ascending=False)
print(salary_gap.to_string(index=False))
```

```python
## q10 Find Top 3 Highest Paid Job Titles per Company Size
df['rank'] = df.groupby('company_size')['salary_in_usd'].rank(method='first', ascending=False)
top3_jobs = df[df['rank'] <= 3].sort_values(['company_size', 'rank'])
print(top3_jobs[['company_size', 'job_title', 'salary_in_usd', 'rank']])
```
